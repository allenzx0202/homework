```python
def factorial2(g):
    if g>0:
        return g*factorial2(g-1)
    else:
        return 1
```

```python
factorial2(5)
```

```python
def fib(n): #write Fibonacci series up to n
    """print a Fibonacci series up to n."""
    a,b=0,1
    while a<n:
        print(a,end=' ')
        a,b=b,a+b
    print()
```

```python
fib(2000)
```

```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('xxx.jpg')
rows,cols,dims=img.shape

img2 = np.uint8(np.zeros((rows,cols,3)))

plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k]
        for k in range(0,3):
            img2[i,j,k] = np.uint8(sum/3)
        
plt.subplot(1,2,2)
plt.title('grey image')
plt.imshow(img2)
plt.show()
```
